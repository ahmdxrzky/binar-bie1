```python
import pandas as pd
from pymongo import MongoClient
```

```python
categories = pd.read_excel("categories.xlsx")
old_columns = categories.columns
columns_dict = {}

for column in old_columns:
    columns_dict[column] = column.lower().replace(" ", "_")
    columns_dict[column] = columns_dict[column].replace("-", "_")

categories = categories.rename(columns=columns_dict)
categories
```

```python
categories.dtypes
```

```python
categories.isnull().sum()
```

```python
categories_no_duplicate = categories['product_id'].drop_duplicates(keep='last')
categories_no_duplicate = categories.loc[categories_no_duplicate.index]
categories_no_duplicate = categories_no_duplicate.reset_index(drop = True)
categories_no_duplicate
```

```python
id_duplicate = list(categories_no_duplicate[categories_no_duplicate['product_id'].duplicated() == True]['product_id'])
# categories_no_duplicate[categories_no_duplicate["product_id"].isin(id_duplicate)]
for id in id_duplicate:
    print(categories_no_duplicate[categories_no_duplicate["product_id"] == id].index)

# categories_no_duplicate = categories_no_duplicate.rename(columns={'product_id': 'prod_id'})
```

```python
data = categories_no_duplicate.to_dict('records')
```

```python
# table_name = "raw_data.categories"

# categories_no_duplicate.to_gbq(
#     table_name,
#     project_id,
#     if_exists="replace",
#     credentials=credentials
# )

client = MongoClient("mongodb://localhost:27017")
db = client['superstore_db']
collection = db['db_products']
collection.insert_many(data)

# categories_no_duplicate.to_csv("products.csv")
```
